# VPD Inputs High Level 

Inputs into VPD are going to look like this 
```text
SportDir
|---pose 
|---crops
\---videos
```



Within the pose directory 
```text 
video_name 
|---boxes.json
|---coco_keypoints.json.gz
|---mask.json.gz
\---meta.json


with each of thes files being formated as
```text 
boxes:
[
    [frame_num, [x, y, w, h]], ...
]

coco_keypoints:
[
    [
        frame_num, [[score, [x, y, w, h], [[x, y, score] * 17]]], ...]
    ],
    ...
]
mask:
[
    [
        frame_num, [[score, [x, y, w, h], base64_encoded_png], ...]
    ],
    ...
]

Meanwhile crops directory will look like 
```text
video_name
|---0.png           // <frame_num>.png
|---0.prev.png
|---0.flow.png
|---0.mask.png
|---1.png
|---1.prev.png
...
```
